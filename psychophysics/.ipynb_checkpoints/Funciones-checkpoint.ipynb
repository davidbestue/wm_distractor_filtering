```python
#functions



# def cw_ccw_detector(target, NT):
#     #print [target, NT,get_quadrant(target),get_quadrant(NT) ]
#     #it returns the relative position of the target respect the NT
#     if target<NT: 
#         if circ_dist(target, NT)<180: o='cw'
#         else: o='ccw'
#     else:
#         if circ_dist(target, NT)<180: o='ccw'
#         else: o='cw'
        
#     #Ecceptions
#     if [get_quadrant(target),get_quadrant(NT) ] == [1,4] : o= 'ccw'
#     if [get_quadrant(target),get_quadrant(NT) ] == [4,1] : o= 'cw'
    
#     return o




def get_quadrant(angle): 
    #angle in degrees, return the quadrant which it belongs to
    if angle>=0 and angle<=90:
        Q=1
    elif angle>90 and angle<=180:
        Q=2
    elif angle>180 and angle<=270:
        Q=3
    elif angle>270 and angle<=360:
        Q=4
    
    return Q



def get_octave(angle): 
    #angle in degrees, return the quadrant which it belongs to
    if angle>=0 and angle<=45:
        Q=1
    elif angle>45 and angle<=90:
        Q=2
    elif angle>90 and angle<=135:
        Q=3
    elif angle>135 and angle<=180:
        Q=4
    elif angle>180 and angle<=225:
        Q=5
    elif angle>225 and angle<=270:
        Q=6
    elif angle>270 and angle<=315:
        Q=7
    elif angle>315 and angle<=360:
        Q=8
    
    return Q



def get_dist_axis(angle):
    dist=min([circ_dist(angle, 0), circ_dist(angle, 90), circ_dist(angle, 180), circ_dist(angle, 270)])
    return dist


def get_dist_ref(angleT, angleR):
    references=[0,45,90, 135, 180, 225, 270, 315]
    opt= [ circ_dist(angleT, 0),  circ_dist(angleT, 45), circ_dist(angleT, 90), 
           circ_dist(angleT, 135), circ_dist(angleT, 180),  circ_dist(angleT, 225),
          circ_dist(angleT, 270),  circ_dist(angleT, 315) ]
    distT=min(opt)
    
    ref = where(array(opt) == distT )[0][0]
    
    distR = circ_dist(angleR, references[ref])
    return distT, distR


    
def get_ref(angleT):
    #same as before, to know the reference we are going
    references=[0,45,90, 135, 180, 225, 270, 315, 360]
    opt= [ circ_dist(angleT, 0),  circ_dist(angleT, 45), circ_dist(angleT, 90), 
           circ_dist(angleT, 135), circ_dist(angleT, 180),  circ_dist(angleT, 225),
          circ_dist(angleT, 270),  circ_dist(angleT, 315) ]
    distT=min(opt)
    
    ref = where(array(opt) == distT )[0][0]
    
    reference= references[ref]
    if reference ==0 :
        if angleT>315:
            reference = 360
        
    return reference


#Con esta función cojo el ángulo y corrijo teniendo en cuenta el modelo 
def correct_bias_axis(angle_T, angle_R, coefs):
    #(angle_t, angle_r, [m, intercept])
    #get the corrected measure according to the model
    correction=coefs[0]*get_dist_axis(angle_T) + coefs[1]
    #correct the response depending on the location
    ref=get_dist_axis(angle_R)
    if get_dist_axis(angle_R-correction) ==  ref - correction:
        return round(angle_R-correction, 3)
    else:
        return round(angle_R+correction, 3)


#Con esta función cojo el ángulo y corrijo teniendo en cuenta el modelo 
def correct_bias_ref(angle_T, angle_R, coefs):
    #(angle_t, angle_r, [m, intercept])
    #get the corrected measure according to the model
    dist_T, dist_R = get_dist_ref(angle_T, angle_R)
    correction=coefs[0]*dist_T + coefs[1]
    #
    #The correction will be cw or ccw deppending on the relative positions of
    # the target and response to the reference (the ref is target based)
    Reference = get_ref(angle_T)
    #print(Reference)
    if (angle_T < Reference) & (angle_R <= Reference):
        return angle_R + correction
    elif (angle_T > Reference) & (angle_R >= Reference):
        return angle_R - correction
    elif (angle_T < Reference) & (angle_R >= Reference):
        return angle_R - correction
    elif (angle_T > Reference) & (angle_R <= Reference):
        return angle_R + correction
    elif angle_T == Reference:
        return angle_R 
    else:
        return 'Error'
        


#Con esta función cojo el ángulo y corrijo teniendo en cuenta el modelo 
def correct_bias_ref2(angle_T, angle_R, coefs):
    #(angle_t, angle_r, [m, intercept])
    #get the corrected measure according to the model
    dist_T, dist_R = get_dist_ref(angle_T, angle_R)
    correction=coefs[0]*dist_T + coefs[1]
    #
    #The correction will be cw or ccw deppending on the relative positions of
    # the target to the reference (the ref is target based)
    Reference = get_ref(angle_T)
    #print(Reference)
    if angle_T < Reference:
        return angle_T + correction
    
    elif angle_T > Reference:
        return angle_T - correction
    
    elif (angle_T == Reference) & (angle_R <= Reference):
        return angle_T + correction
    
    elif (angle_T == Reference) & (angle_R >= Reference):
        return angle_T - correction

    

    
        
#Con esta función cojo el ángulo y corrijo teniendo en cuenta el modelo 
def correct_bias_ref3(angle_T, angle_R, coefs):
    #(angle_t, angle_r, [m, intercept])
    #get the corrected measure according to the model
    dist_T, dist_R = get_dist_ref(angle_T, angle_R)
    correction=coefs[0]*dist_T + coefs[1]
    #
    #The correction will be cw or ccw deppending on the relative positions of
    # the target to the reference (the ref is target based)
    Reference = get_ref(angle_T)
    #print(Reference)
    if (angle_T < Reference) & (angle_R <= Reference):
        return angle_T + correction
    elif (angle_T > Reference) & (angle_R >= Reference):
        return angle_T - correction
    elif (angle_T < Reference) & (angle_R >= Reference):
        return angle_T - correction
    elif (angle_T > Reference) & (angle_R <= Reference):
        return angle_T + correction
    elif angle_T == Reference:
        return angle_T 




#### functions

def subject_number(code_subj):
    if code_subj == 'TH':
        index=1
    elif code_subj == 'AZ':
        index=2
    elif code_subj == 'ER':
        index=3
    elif code_subj == 'AN':
        index=4
    elif code_subj == 'HL':
        index=5
    elif code_subj == 'LM':
        index=6        
    elif code_subj == 'MI':
        index=7 
    elif code_subj == 'TE':
        index=8
    elif code_subj == 'LI':
        index=9
    elif code_subj == 'FL':
        index=10
    elif code_subj == 'YH':
        index=11
    elif code_subj == 'AR':
        index=12
    elif code_subj == 'SW':
        index=13
    elif code_subj == 'HM':
        index=14
    elif code_subj == 'HT':
        index=15
    elif code_subj == 'FE':
        index=16
    elif code_subj == 'LA':
        index=17
    elif code_subj == 'ML':
        index=18
    elif code_subj == 'MH':
        index=19
    elif code_subj == 'SR':
        index=20
    elif code_subj == 'SN':
        index=21
    elif code_subj == 'n001':
        index=22
    elif code_subj == 'b001':
        index=23    
    elif code_subj == 'l001':
        index=24 
    elif code_subj == 's001':
        index=25 
    elif code_subj == 'r001':
        index=26 
    elif code_subj == 'd001':
        index=27 
    elif code_subj == 'a001':
        index=35 
    else:
        index=99999
    
    
    return index




def male_female(code):
    if code in [2,5,15,16,19]:
        sex = 'M'
    else:
        sex='F'
    
    return sex




def rgb_norm(a,b,c):
	return [float(a)/255, float(b)/255, float(c)/255]
 
 



def circ_dist(a1,a2):
    ## Returns the minimal distance in angles between to angles 
    op1=abs(a2-a1)
    angs=[a1,a2]
    op2=min(angs)+(360-max(angs))
    options=[op1,op2]
    return min(options)






def give_closer_NT(T,NT1,NT2):
    dist_NT1=[circ_dist(T[i], NT1[i]) for i in range(0, len(T))]
    dist_NT2=[circ_dist(T[i], NT2[i]) for i in range(0, len(T))]
    list_index=zeros((len(T)))    
    for i in range(0, len(T)):
        if dist_NT1[i]<dist_NT2[i]:
            list_index[i]=NT1[i]
        else:
            list_index[i]=NT2[i]
    
    return list_index





def give_closer_dist(T,dist1, dist2, dist3):
    dist_dist1=[circ_dist(T[i], dist1[i]) for i in range(0, len(T))]
    dist_dist2=[circ_dist(T[i], dist2[i]) for i in range(0, len(T))]
    dist_dist3=[circ_dist(T[i], dist3[i]) for i in range(0, len(T))]
    
    list_index=zeros((len(T), 3))    
    for j in range(0, len(T)):
        m=zeros((3,2))
        m[0,0]=dist_dist1[j]
        m[1,0]=dist_dist2[j]
        m[2,0]=dist_dist3[j]
        m[:,1]=[1,2,3]
        m_sort=m[m[:,0].argsort()]
        
        list_index[j,0]=m_sort[0,1]
        list_index[j,1]=m_sort[1,1]
        list_index[j,2]=m_sort[2,1]
    
    
    return list_index





def new_A_resp(target, close, bias):
    # put the actual loctaions and give the bias to correct the response (the bias must be the err interference with the close NT)
    if bias>0:
        #me tendre que acercar al close
        dist_actual=circ_dist(target, close)
        if circ_dist(target+bias, close) < dist_actual:
            new_a_r=target+bias
        else:
            new_a_r=target-bias
    
    else:
        #me tendre que alejar
        dist_actual=circ_dist(target, close)
        if circ_dist(target+bias, close) > dist_actual:
            new_a_r=target+bias
        else:
            new_a_r=target-bias
    
    
    return new_a_r



def give_closer_dist(T,dist1, dist2, dist3):
    dist_dist1=[circ_dist(T[i], dist1[i]) for i in range(0, len(T))]
    dist_dist2=[circ_dist(T[i], dist2[i]) for i in range(0, len(T))]
    dist_dist3=[circ_dist(T[i], dist3[i]) for i in range(0, len(T))]
    
    list_index=zeros((len(T), 3))    
    for j in range(0, len(T)):
        m=zeros((3,2))
        m[0,0]=dist_dist1[j]
        m[1,0]=dist_dist2[j]
        m[2,0]=dist_dist3[j]
        m[:,1]=[1,2,3]
        m_sort=m[m[:,0].argsort()]
        
        list_index[j,0]=m_sort[0,1]
        list_index[j,1]=m_sort[1,1]
        list_index[j,2]=m_sort[2,1]
    
    
    return list_index
    

    

# def give_closer_dist_angle(T,dist1, dist2, dist3):
#     closest=[]
#     for i in range(0, len(T)):
#         options = [circ_dist(T[i], dist1[i]),circ_dist(T[i], dist2[i]), circ_dist(T[i], dist3[i]) ]
#         ind = where(options == min(options))[0][0]
#         #print ind
#         closest.append( [dist1[i], dist2[i], dist3[i]][ind] )

#     return closest


def give_closer_dist_angle(T,dist1, dist2, dist3):
    closest=[]
    for i in range(0, len(T)):
        options = [circ_dist(T[i], dist1[i]),circ_dist(T[i], dist2[i]), circ_dist(T[i], dist3[i]) ]
        #print(options)
        ind = where(array(options) == min(options))[0][0]
        #print(ind)
        closest.append( [dist1[i], dist2[i], dist3[i]][ind] )

    return closest




def bootstrap(data, num_samples, statistic, alpha):
    """Returns bootstrap estimate of 100.0*(1-alpha) CI for statistic."""
    n = len(data)
    idx = npr.randint(0, n, (num_samples, n))
    samples = data[idx]
    stat = np.sort(statistic(samples, 1))
    return (stat[int((alpha/2.0)*num_samples)],
            stat[int((1-alpha/2.0)*num_samples)])






def significance_bar(x1,x2, y1, y2, pval, orientation, linewidth = 1.2,markersize = 8,boxpad  =0.3,fontsize = 10,color = 'k'):
    if orientation=='left':
        marker=TICKRIGHT
    elif orientation=='right':
        marker=TICKLEFT
    else:
        marker=TICKDOWN
    
    if pval>=0.05:
        displaystring = r'ns'
    elif pval<0.0001:
        displaystring = r'***'
    elif pval<0.001:
        displaystring = r'**'
    else:
        displaystring = r'*'
    # draw a line with downticks at the ends
    plt.plot([x1,x2],[y1,y2],'-',color = color,lw=linewidth,marker = marker,markeredgewidth=linewidth,markersize = markersize)
    # draw the text with a bounding box covering up the line
    if orientation=='left':
        plt.text(x1,0.5*(y1+y2), displaystring,ha = 'center',va='center',bbox=dict(facecolor='1.', edgecolor='none',boxstyle='Square,pad='+str(boxpad)),size = fontsize)
    elif orientation=='right':
        plt.text(x1,0.5*(y1+y2), displaystring,ha = 'center',va='center',bbox=dict(facecolor='1.', edgecolor='none',boxstyle='Square,pad='+str(boxpad)),size = fontsize)
    else:
        plt.text(0.5*(x1+x2), y1, displaystring,ha = 'center',va='center',bbox=dict(facecolor='1.', edgecolor='none',boxstyle='Square,pad='+str(boxpad)),size = fontsize)









#quadrant

def quadrant_1(a):
    q_1=a[:,4]<90
    return(q_1)


def quadrant_2(a):
    q_2a=a[:,4]>90
    q_2b=a[:,4]<180
    q_2=q_2a & q_2b
    return(q_2)


def quadrant_3(a):
    q_3a=a[:,4]>180
    q_3b=a[:,4]<270
    q_3=q_3a & q_3b
    return(q_3)


def quadrant_4(a):
    q_4=a[:,4]>270
    return(q_4)


def quadrant(a, b):
    if b==1:
        q=quadrant_1(a)
        return(q)
    elif b==2:
        q=quadrant_2(a)
        return(q)
    elif b==3:
        q=quadrant_3(a)
        return(q)
    elif b==4:
        q=quadrant_4(a)
        return(q)




##### Orientation

def cw(a):
    cw=a[:,20]==1
    return(cw)


def ccw(a):
    ccw=a[:,20]==-1
    return(ccw)


def cw_ccw_disc(a, orient): #orient 1 is cw and orient 2 is ccw, the reason is to make easy to append
    if orient==1:
        m=cw(a)
        return(m)
    elif orient==2:
        m=ccw(a)
        return(m)    







def reg_analysis(Y, X, X_header='Default', summary='True'): 
    
    #Y and X must be np arrays
    num_parameters=shape(X)[1]
    if X_header=='Default':
        X_header=['Intercept']
        for i in range(1, num_parameters):
            X_header.append(str(range(1, num_parameters)[i-1]))
    
    if shape(X_header)[0]!=num_parameters:
        return 'Add the intercept to your X_header? No match between columns and headers'
    
    
    
    #Your X matrix must include a first column of 1 (intercept)
    
    # Y = beta*X + error
    # X'Y = (X'X) betas
    ##############################################################
    ##############################################################
    #The folowing is the most important one: (the one we are going to develope)
    
    # inv(X' X) * (X' Y) = betas
    
    ##############################################################
    ############################################################## 
    
    #I get the tranpose of X
    X_Xt=np.dot(X.T,X)
    
    try:
        X_Xt_inv=np.linalg.inv(X_Xt)
    except:
        return "Is it a singular matrix? Be carefull with categorical variables" 
    
    
    if round(np.linalg.det(X_Xt), 5)==0:
        print('It passed the inversion but it should not, there is a duplicated column')
    
    
    #I do this operation (X' Y)
    X_t_Y=np.dot(X.T,Y)
    
    betas=np.dot(X_Xt_inv , X_t_Y)
    
    ##########
    # E = Y - X* betas  
    errors= Y - np.dot(X, betas)
    
    # etsimated= X* betas
    estimated_values=np.dot(X, betas)
    
    #var = E'E / N - parameters
    # def freedom: number subjects - parameters
    N=shape(X)[0]
    deg_freedom=N - num_parameters
    variance_errors= np.dot(errors.T, errors) / deg_freedom
    
    # Variance betas
    # var(betas) = variance_errors * inv(X' X) 
    #I extract the diagonal because is a p, p matrix!! DUDO!!!???
    
    beta_variances=np.diag(variance_errors*X_Xt_inv)
    beta_stds= [sqrt(beta_variances[i]) for i in range(0, len(beta_variances))]
    
    #p_values 
    ##
    # Interpretation
    # http://stats.stackexchange.com/questions/128723/understanding-of-p-value-in-multiple-linear-regression
    #
    #The p-value for each term tests the null hypothesis that the coefficient is equal to zero (no effect). 
    #A low p-value (< 0.05) indicates that you can reject the null hypothesis. In other words, a predictor that 
    #has a low p-value is likely to be a meaningful addition to your model because changes in the predictor's 
    #value are related to changes in the response variable.
    #
    t_values=betas/beta_stds
    #
    # p values from a t distribution. As it is two tailed I have to *2 the p value (regressors can be negative)
    p_values=t.sf(abs(t_values), deg_freedom)*2
    
    ## dictionary with results
    results={}
    results['coeficients']=betas
    results['beta_variances']=beta_variances
    results['beta_stds']=beta_stds
    results['deg_freedom']=deg_freedom
    results['num_parameters']=num_parameters
    results['N']=N
    results['errors']=errors
    results['estimated_values']=estimated_values
    results['variance_errors']=variance_errors
    results['t_values']= t_values
    results['p_values']= p_values
    
    
    #display the table if wanted
    # https://pypi.python.org/pypi/tabulate 
    if summary=='True':
        decimals=3
        keys_gen=['Model  ', 'N', 'num params', 'df_residuals', 'df_model', 'var errors' ] 
        data_gen=[[ 'OLS Regression', N, num_parameters, deg_freedom, num_parameters-1, round(variance_errors, decimals) ]]
        
        keys_stat=['parameter', 'coef', 'std err', 't', 'p_value' ]
        data_stat=[[] for i in range(0, len(betas))]
        
        for i in range(0, len(betas)):
            data_stat[i]=[X_header[i], round(betas[i], decimals), round(beta_stds[i], decimals), round(t_values[i], decimals), round(p_values[i], decimals)]
        
        print( '                                                     ')
        print( '                                                     ')
        print( '                  Summary table                      ')
        print( '-----------------------------------------------------')
        print( '                                                     ')
        print( '                  General data                       ')
        print( tabulate(data_gen, headers=keys_gen, tablefmt="simple"))    
        print( '                                                     ')    
        print( '                   Statistics                        ')
        print( tabulate(data_stat, headers=keys_stat, tablefmt="rst"))
        
        return results
    
    else:
        return results
    
    

#############################
############################
#function interactions


def matrix_interactions( X, indices, X_header=[] ):
    #indices= [ [], [] ] one claudator for each interaction
    # compatible both with the headers and the index of the columns
    dictionary={}
       
    if X_header!=[]:        
        for i in range(0, len(X_header)):
            dictionary[X_header[i]]=i
    else:
        for i in range(0, shape(X)[1]):
            dictionary[str(i)]=i    
    
    
    # create the matrix with the extracolumns      
    M=np.zeros((   len(X),    len(indices) ))
    
    #fill this extra columns with the interactions and generate the new header
    for i in range(0, len(indices)):
        M[:,  i ] = list(  X[:, dictionary[indices[i][0]] ] *  X[:, dictionary[indices[i][1]] ] )
        if X_header!=[]:
            X_header.append( indices[i][0] + ' * ' + indices[i][1] )
    
    #put together both matrix
    Matrix=np.hstack(( X, M )) 
    
    return Matrix









def matrix_interactions( X, indices, X_header=[] ):
    #indices= [ [], [] ] one claudator for each interaction
    # compatible both with the headers and the index of the columns
    dictionary={}
       
    if X_header!=[]:        
        for i in range(0, len(X_header)):
            dictionary[X_header[i]]=i
    else:
        for i in range(0, shape(X)[1]):
            dictionary[str(i)]=i    
    
    
    # create the matrix with the extracolumns      
    M=np.zeros((   len(X),    len(indices) ))
    
    #fill this extra columns with the interactions and generate the new header
    for i in range(0, len(indices)):
        if len(indices[i])==2:
            M[:,  i ] = list(  X[:, dictionary[indices[i][0]] ] *  X[:, dictionary[indices[i][1]] ] )
            if X_header!=[]:
                X_header.append( indices[i][0] + ' * ' + indices[i][1] )
        if len(indices[i])==3:
            M[:,  i ] = list(  X[:, dictionary[indices[i][0]] ] *  X[:, dictionary[indices[i][1]] ] * X[:, dictionary[indices[i][2]] ] )
            if X_header!=[]:
                X_header.append( indices[i][0] + ' * ' + indices[i][1] + ' * ' + indices[i][2] )    
    #put together both matrix
    Matrix=np.hstack(( X, M )) 
    
    return Matrix













#function intercept

def add_intercept(X, X_header=[]):
    #Add one column for the intercept and add the string intercept to the keys
    if X_header!=[]:
        X_header.insert(0, 'Intercept')
    
    M=np.ones((   len(X),    1 ))
    Matrix=np.hstack(( M, X )) 
    return Matrix






def bootstrap_subj_samematrix(DATA, col_int, indx_subj, num_samples, statistic, alpha):  #here takes all the data of one subject, not all the data together.
    """Returns bootstrap estimate of 100.0*(1-alpha) CI for statistic."""
    dat=DATA[:, col_int]
    data=DATA
    n = len(unique(DATA[:, indx_subj]))
    idx = random.randint(0, n, (num_samples, n))
    ##
    pool=[[] for i in range(0, num_samples)]
    for i, IDX in enumerate(idx):
        index=IDX
        for subj in index:
            pool[i].append(dat[data[:, indx_subj]==subj])

    samples=[list(chain.from_iterable(pool[i])) for i in range(0, len(pool))]
    ##
    r=[statistic(samples[i]) for i in range(0, len(samples))]  
    r=list(r)
    r.sort()
    stat=array(r)
    return (stat[int((alpha/2.0)*num_samples)],
            stat[int((1-alpha/2.0)*num_samples)])


    
    









    



```

```python
def circ_dist(a1,a2):
    ## Returns the minimal distance in angles between to angles 
    op1=abs(a2-a1)
    angs=[a1,a2]
    op2=min(angs)+(360-max(angs))
    options=[op1,op2]
    return min(options)



def angle_error(T,R):
    error = T - R
    if abs(error) == circ_dist(T,R):
        return error
    
    #When one is in the first and the other in the 4th quadrant
    else:
        if T<R:
            return circ_dist(T,R)
        
        else:
            return -circ_dist(T,R)



def angle_error_col(x): #T,R
    T = x[0]
    R=  x[1]
    
    error = T - R
    if abs(error) == circ_dist(T,R):
        return error
    
    #When one is in the first and the other in the 4th quadrant
    else:
        if T<R:
            return round(circ_dist(T,R), 3)
        
        else:
            return round(-circ_dist(T,R), 3)



def Interference_effects(target, response, reference):
    #input list of target, list of responses and list of references
    #Error_interference; positive for attraction and negative for repulsion
    #######
    #Decimals to get
    decimals=3
    ####
    interferences=[]
    for i in range(0, len(target)):
        angle_err_abs=abs(circ_dist(target[i],response[i]))
        if circ_dist(array(response)[i], array(reference)[i])<=circ_dist(array(target)[i], array(reference)[i]):
            Err_interference=round( angle_err_abs, decimals) 
        else:
            Err_interference=round( -angle_err_abs, decimals)

        interferences.append(Err_interference)
    
    return interferences
```

```python
def circdist(array1,array2):
    from numpy import array
    res=[]
    ## Returns the minimal distance in angles between to arrays of angles
    for i in range(0, len(array1)):
        op1=abs(array2[i]-array1[i])
        angs=[array1[i],array2[i]]
        op2=min(angs)+(360-max(angs))
        options=[op1,op2]
        res.append( min(options) )
    
    return array(res)
```

```python
def dotplot(df, columns, subjects, dotsize=10, dotsizemin=5, colors='g', xnames='',
             title='', xlabel='', ylabel=''):
    #######################################################################################
    #######################################################################################
    import numpy as np
    import numpy.random as npr
    from numpy import array, mean, random
    import matplotlib.pyplot as plt
    from scikits.bootstrap import ci 
    
    #Change to list in case it is just one
    if isinstance(columns, str)==True:
        columns=[columns]
        xnames =[xnames]
    
    if isinstance(colors, str)==True:
        colors=[colors]
    
    #Function
    def bootstrap_subjlevel(df, col_int, indx_subj):
        data_grouped=df.groupby(indx_subj)
        sup, inf =     ci([data_grouped.get_group(i)[col_int].mean() for i in df[indx_subj].unique()])
        return sup, inf
    
    ########################################################################################
    ########################################################################################
    #Plot for each column
    for xpos, column in enumerate(columns):
        #Color
        if len(colors)>1:
            color=colors[xpos]
        else:
            color = colors[0]
        
        #Confidence interval subject
        sup, inf = bootstrap_subjlevel(df, column, subjects)
        errors=(   [abs(sup -df[column].mean())], [abs(df[column].mean() - inf)])

        #Plot
        plt.errorbar(xpos, df[column].mean(), yerr=errors, ecolor='k', alpha=0.4, fmt='o', capsize=2)
        plt.plot(xpos, df[column].mean(), color=color,marker = 'o', markersize=dotsize, alpha=1 )
        A=df.groupby([subjects])
        Means = [A.get_group(i)[column].mean() for i in list(df.subject.unique())]
        [plt.plot(xpos, Means[i], color=color, marker = 'o', markersize=dotsizemin, alpha=0.4 ) for i in range(len(Means))]
        ###
    
    #xticks
    if xnames!='':
        plt.xticks(list(range(0, len(columns))), xnames) 
    
    #features
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.title(title) if title!='' else plt.title('')
    plt.xlabel(xlabel) if title!='' else plt.xlabel('')
    plt.ylabel(ylabel) if title!='' else plt.ylabel('')
    plt.xlim(-3, xpos+3)
    plt.show(block=False)
        

```
